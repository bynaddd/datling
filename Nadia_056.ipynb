```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

from selenium import webdriver
import chromedriver_autoinstaller
```

```python
#install browser such chrome or edge
chromedriver_autoinstaller.install()
# use the installed from sellenium
browser= webdriver.Chrome()
```

```python
browser.get('https://www.sociolla.com/promo/7679-bestie-wishlists-on-sale?page=1&limit=50')
soup= BeautifulSoup(browser.page_source,'lxml')
```

```python
rows = soup.find_all('div',class_= 'product product--v3')
```

```python
len(rows)
```

```python
data=[]
for row in rows:
    diskon_produk=row.find(class_='product__info-sale product__info-sale--percentage')
    diskon = diskon_produk.text if diskon_produk is not None else ""
    nama_produk=row.find(class_='product__brand')
    merek = nama_produk.text if nama_produk is not None else ""
    deskripsi_produk=row.find(class_='product__name')
    deskripsi = deskripsi_produk.text if deskripsi_produk is not None else ""
    hargaseb_produk=row.find(class_='product-price__stroke')
    harga_normal= hargaseb_produk.text if hargaseb_produk is not None else ""
    riview_produk=row.find(class_='product__reviews')
    telah_diriview_sebanyak = riview_produk.text if riview_produk is not None else ""

    dt = {
        'diskon' : diskon,
        'merek' : merek,
        'deskripsi' : deskripsi,
        'harga_normal' : harga_normal,
        'telah_diriview_sebanyak': telah_diriview_sebanyak
    }
    data.append(dt)
```

```python
df = pd.DataFrame(data)
```
